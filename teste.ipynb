```python
import datetime

# from tables import Column
import time

import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np
import pandas as pd
import scipy.stats as stats
import yfinance as yf
```

```python
def FormataStringData(data):

  dia  = data.split("/")[0]
  mes  = data.split("/")[1]
  ano  = data.split("/")[2]

  return ano + '-' + mes + '-' + dia
```

```python
stock_list = ['ITUB3.SA', 'VALE3.SA']

df_stock = pd.DataFrame()

for stock in stock_list:
    df_stock[stock] = yf.download(stock, start=FormataStringData('01/12/2022'), end=FormataStringData('01/01/2023'))["Adj Close"]
    
df_stock.reset_index(inplace=True)
df_stock = df_stock.melt(id_vars='Date', var_name='Tickers', value_name='Adj_Close')
```
